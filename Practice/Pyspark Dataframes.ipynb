```python
# how==any - if the row has even one NULL value
df2=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')
df2.na.drop(how='any').show()
```

```python
#how==all -if entire row has NULL values
df2=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')
df2.na.drop(how='all').show()
```

```python
#threshold = atleast (thershold) number of Non-NULL values must be present in row

df2=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')
df2.na.drop(how="any",thresh=2).show()

```

```python
# subset - drops the rows from mentioned column
df2.na.drop(how="any",subset=["Name"]).show()
```

```python
# fill the missing values - fills the Nan values with mentioned value
df2=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')
df2.na.fill("Missing value").show()
```

```python
# fill the missing values - fills the Nan values with mentioned value, fro particular column
df2=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')
df2.na.fill("Missing value",'Age').show()
```

```python
# handling missing value by filling the Nan values

from pyspark.ml.feature import Imputer
df3=spark.read.option('header','true').csv('/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv',inferSchema='true')

imputer = Imputer(
    inputCols=['Age','Experience','Salary'],
    outputCols=["{}_imputed".format(c) for c in ['Age','Experience','Salary']]
    ).setStrategy("mean")

imputer.fit(df3).transform(df3).show()
```
