```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("Simple Pipeline").getOrCreate()
df=spark.read.csv("/Workspace/Repos/pusthakaloka@gmail.com/Databricks/Practice/testdata2.csv", header=True, inferSchema=True)

#clean
df_clean=df.dropna()

#transform
df_filtered=df_clean.filter(df_clean.Salary>20000)
df_grouped=df_filtered.groupBy("Salary").count()

df_grouped.show()
```
